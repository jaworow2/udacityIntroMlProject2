````python
# TODO: Get the number of examples in each set from the dataset info.
num_train_set = dataset_info.splits["train"].num_examples
num_val_set = dataset_info.splits["validation"].num_examples
num_test_set = dataset_info.splits["test"].num_examples

print("Number of training set examples: ", num_train_set)
print("Number of validation set examples: ", num_val_set)
print("Number of test set examples: ", num_test_set)

# TODO: Get the number of classes in the dataset from the dataset info.
num_classes = dataset_info.features["label"].num_classes

print("Number of classes: ", num_classes)
````

````python
def image_label(train, img_cnt):
    '''
    Loops three tensorflow training set returning and provides exploratory 
    data
    
    arguments:
        train: tensorflow image training set
        imge_cnt: number of images
    
    returns:
        returns three lists: image, label, and shape  of the tensorflow 
        image
    '''
    img = []
    lab = []
    shap = []
    for image, label in train.take(img_cnt):
        img.append(image.numpy())
        shap.append(image.shape)
        lab.append(label.numpy())
    return img, lab, shap
````

## Create Pipeline

````python
# TODO: Create a pipeline for each set.
batch_size = 32
image_size = 224

def format_image(image, label, augment=False):
    '''
    Formats the tensor image by changing to a tensor float data type,
    setting a uniform size, setting values between 0 and 1, and if a training 
    set applying random transformations
    
    arguments:
        image: tensorflow image
        label: label of tensorflow image
        augment: indicates whether transformations will be applied
    
    returns:
        returns formatted tensorflow image and label

    Code Attribution:
        This function contains code that was updated and leveraged from the 
        Udacity Intro to Machine Learning Nano Degree Training material.
    '''
    image = tf.cast(image, tf.float32)
    image = tf.image.resize(image, (image_size, image_size))
    image /= 255
    
    # apply image augmentation if true
    if augment==True:
        image = tf.image.random_flip_left_right(image, seed=42)
        image = tf.image.random_flip_up_down(image, seed=42)
        image = tf.image.random_jpeg_quality(image, 90, 95, seed=42)
    return image, label

training_batches = (training_set
                       .cache()
                       .shuffle(num_train_set//4)   
                       .padded_batch(batch_size)
                       .map(format_image, True)
                       .prefetch(1)
                    )

validation_batches = (validation_set
                          .cache()
                          .padded_batch(batch_size)
                          .map(format_image)
                          .prefetch(1)
                     )

testing_batches = (test_set.cache()
                       .padded_batch(batch_size)
                       .map(format_image)
                       .prefetch(1)
                  )
````

````python
# TODO: Build and train your network.

# Pre-trained MobileNet URL
URL = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"

# retrieve the MobileNet pre-trained model from TensorFlow Hub
feature_extractor = hub.KerasLayer(URL, input_shape=(image_size, image_size,3))

# do not train the pre-train model
feature_extractor.trainable = False
````

````python
# build the sequential model with the pre-trained MobileNet as the feature extractor
tf.keras.backend.clear_session()

model = tf.keras.Sequential()
model.add(feature_extractor)
model.add(tf.keras.layers.Dense(500, activation = 'relu'))
model.add(tf.keras.layers.Dropout(.2))
model.add(tf.keras.layers.Dense(102, activation="softmax"))

model.summary()
````

````python
# set the models optimizer
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
````

````python
EPOCHS = 10

early_stopping = tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=2)

history = model.fit(training_batches,
                    epochs=EPOCHS,
                    validation_data=validation_batches,
                    callbacks=[early_stopping])
````

````python
# TODO: Plot the loss and accuracy values achieved during training for the training and validation set.
def display_loss_accuracy(history):
    ''' 
    Receives a tensorflow model history object and 
    displays training and testing accuracy and 
    loss line graphs
    
    argument:
        history: tensorflow model history object
    
    returns:
        none
        
    Code Attribution:
        This function contains code that was leveraged from the 
        Udacity Intro to Machine Learning Nano Degree Training material.
    '''
    # training and validation accuracy from tensorflow model object
    training_accuracy = history.history['accuracy']
    validation_accuracy = history.history['val_accuracy']

    # training and validation loss from tensorflow model object
    training_loss = history.history['loss']
    validation_loss = history.history['val_loss']

    # range of the number of epochs the model ran
    epochs_range=range(len(training_accuracy))

    # build the training and validation accuracy graph
    plt.figure(figsize=(8, 8))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, training_accuracy, label='Training Accuracy')
    plt.plot(epochs_range, validation_accuracy, label='Validation Accuracy')
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    # build the training and validation loss graph
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, training_loss, label='Training Loss')
    plt.plot(epochs_range, validation_loss, label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.show()

display_loss_accuracy(history)
````

### Image Pre-processing

The `process_image` function should take in an image (in the form of a NumPy array) and return an image in the form of a NumPy array with shape `(224, 224, 3)`.

First, you should convert your image into a TensorFlow Tensor and then resize it to the appropriate size using `tf.image.resize`.

Second, the pixel values of the input images are typically encoded as integers in the range 0-255, but the model expects the pixel values to be floats in the range 0-1. Therefore, you'll also need to normalize the pixel values. 

Finally, convert your image back to a NumPy array using the `.numpy()` method.

````python
# TODO: Create the process_image function
def process_image(image):
    ''' 
    Converts, resizes, and normalizes a NumPy array image into a 
    uniform shape
    
    argument:
        image: NumPy array image with varying shape (varies, varies, 3)
    
    returns:
        image: Numpy array with shape (224, 224, 3)
    
    Code Attribution:
        This function contains code that was updated and leveraged from the 
        Udacity Intro to Machine Learning Nano Degree Training material.
    '''
    image_size = 224
    
    # convert image into a TensorFlow Tensor
    image_file = tf.convert_to_tensor(image)
    
    # resize to the appropriate size
    image_file = tf.image.resize(image_file, (image_size, image_size))
    
    # normalize the pixels
    image_file /= 255
    
    # convert into NumPy array
    image = image_file.numpy()
    
    return image
````

### Inference

Remember, the `predict` function should take an image, a model, and then returns the top $K$ most likely class labels along with the probabilities. The function call should look like:

Your `predict` function should use `PIL` to load the image from the given `image_path`. You can use the [Image.open](https://pillow.readthedocs.io/en/latest/reference/Image.html#PIL.Image.open) function to load the images. The `Image.open()` function returns an `Image` object. You can convert this `Image` object to a NumPy array by using the `np.asarray()` function.

**Note:** The image returned by the `process_image` function is a NumPy array with shape `(224, 224, 3)` but the model expects the input images to be of shape `(1, 224, 224, 3)`. This extra dimension represents the batch size. We suggest you use the `np.expand_dims()` function to add the extra dimension. 

````python
# TODO: Create the predict function
def predict(image_path, model, top_k):
    ''' 
    Takes a passed image, trained mode, and desired top 
    probabilities and returns prediction probabilities 
    for each image class
    
    argument:
        image_path: image to be predicted
        mode: trained model
        top_k: number of top prediction probabilities
    
    returns:
        class_prob: predicted probability for image classes
        class_index: predicted class index
    
    Code Attribution:
        This function contains code that was updated and leveraged from the 
        Udacity Intro to Machine Learning Nano Degree Training material.
    '''
    # load image
    im = Image.open(image_path)
    
    # convert to NumPy array
    test_image = np.asarray(im)
    
    # reshape Numpy array with shape (224, 224, 3)
    processed_test_image = process_image(test_image)
    
    # add extra dimension (1, 224, 224, 3) for the required model format
    processed_test_image = np.expand_dims(processed_test_image, axis=0)

    # numpy array 1x102 of the probability prediction value for each flower image
    image_prediction = model.predict(processed_test_image)
    
    # tensorflow method to identify a tensor of the top K predicted classes 
    # probabilities and indices
    class_prob, class_index = tf.math.top_k(image_prediction, top_k) 
    
    # convert predicted class tensor to numpy array list
    class_prob = class_prob.numpy().tolist() 

    # convert predicted class index tensor to numpy array list
    class_index = class_index.numpy().tolist()
    
    # add 1 since index starts at 0
    class_index = [i+1 for i in class_index[0]]
    
    return class_prob[0], class_index
````

You can convert from the class integer labels to actual flower names using `class_names`. 

````python
def sorted_class_prob(class_names, classes, probs, ascend):
    ''' 
    Takes a dictionary of flower image class_names, predicted
    flower image index, predicted probability, and sort preference 
    and returens a sorted dataframe
    
    argument:
        class_names: dictionary of flower images
        classes: predicted flower indexes
        probs: predicted flower probabilities
        ascend: sorted direction
    
    returns:
        df_sorted: a dataframe of sorted indexes and classes
    '''
    class_display = [class_names[str(class_index)] for class_index in classes]
    df = pd.DataFrame(
        {"class_name" : class_display,
         "prob" : probs})
    df_sorted = df.sort_values("prob", ascending=ascend)
    return df_sorted
````
